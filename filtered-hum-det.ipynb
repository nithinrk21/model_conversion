```python
from ultralytics import YOLO

# Load a base pretrained YOLO model (YOLOv8n here â€” fast and lightweight)
model = YOLO("yolo11n.pt")

results=model.train(
    data="/kaggle/input/filtered-human/data.yaml",
    epochs=100,
    imgsz=640,
    batch=8,
    patience=20,
    cos_lr=True,
    fliplr=0.5,
    hsv_h=0.005,
    hsv_s=0.1,
    hsv_v=0.1,
    verbose=True,
    project="runs",
    name="yolo11n-humanonly-clean",
    exist_ok=True
)
```
